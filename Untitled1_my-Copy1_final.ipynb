```python
# Defining the remove_channels dictionary
remove_channels = {
    "batch1": [8, 10, 12, 13, 22],
    "batch2": [53, 54, 55, 61, 62],
    "batch3": [96, 117, 126, 131, 136, 137]
}
# Creating the list of channels to be removed based on the provided dictionary
remove_files = []
for batch, channels in remove_channels.items():
    for channel in channels:
        remove_files.append(f"Severson_{batch}_channel{channel}_telemetry_cycle10.csv")
        remove_files.append(f"Severson_{batch}_channel{channel}_telemetry_cycle100.csv")

# Filtering out the specified channels
filtered_files = [file for file in files if file not in remove_files]

len(filtered_files)  # Verify we have 248 files after removal
```

```python
# Containers to store loaded data
all_data_cycle10 = {}
all_data_cycle100 = {}

# Iterating through the filtered files and loading data for each channel
for csv_file in filtered_files:
    channel_num = int(csv_file.split("channel")[1].split("_")[0])  # Extracting channel number from filename
    cycle_num = int(csv_file.split("cycle")[1].split(".")[0])  # Extracting cycle number from filename
    
    if cycle_num == 10:
        all_data_cycle10[channel_num] = pd.read_csv(os.path.join(extract_folder, csv_file))
    elif cycle_num == 100:
        all_data_cycle100[channel_num] = pd.read_csv(os.path.join(extract_folder, csv_file))

# Displaying the first few rows of the loaded data for channel0 to ensure it's loaded correctly
all_data_cycle10[0].head(), all_data_cycle100[0].head()
```

```python

# Function to clean and return voltage and discharge values
def discharge_curve(vs, qds, clean=True):
    if not clean:
        return vs, qds
    start = np.argmax(vs)
    end = np.argmin(vs[start:]) + start
    vs = vs[start:end]
    qds = qds[start:end]
    return vs, qds





# Function to overlay discharge curves for all channels on a single plot
def overlay_discharge_curves_for_all_channels(cycle10_data, cycle100_data):
   

    # Plotting discharge curves for cycle 10
    for channel_num, data in cycle10_data.items():
        voltage, discharge = discharge_curve(data["u_bat_raw"].values, data["ca_disch"].values, clean=True)
        plt.plot(discharge, voltage, color='blue', alpha=0.5)

    # Plotting discharge curves for cycle 100
    for channel_num, data in cycle100_data.items():
        voltage, discharge = discharge_curve(data["u_bat_raw"].values, data["ca_disch"].values, clean=True)
        plt.plot(discharge, voltage,  color='red', alpha=0.5)

    # Setting plot details
    plt.title('Overlay of Discharge Curves for All Channels',fontsize=20)
    plt.xlabel('Discharge (Ah)',fontsize=16)
    plt.ylabel('Voltage (V)',fontsize=16)
    plt.legend(['Cycle 10', 'Cycle 100'])
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Plotting the overlay of discharge curves for all channels
overlay_discharge_curves_for_all_channels(all_data_cycle10, all_data_cycle100)
```

```python
# Adapting the code to merge and manipulate data structures for all channels

# Placeholder for the merged data structure
all_batch_adapted = {}

# Merging the data
for channel_num, data_cycle10 in all_data_cycle10.items():
    data_cycle100 = all_data_cycle100[channel_num]
    voltage_10, discharge_10 = discharge_curve(data_cycle10["u_bat_raw"].values, data_cycle10["ca_disch"].values, clean=True)
    voltage_100, discharge_100 = discharge_curve(data_cycle100["u_bat_raw"].values, data_cycle100["ca_disch"].values, clean=True)
    
    # Storing the data in the new structure
    all_batch_adapted[channel_num] = {
        'cycle10': {
            'voltage': voltage_10,
            'discharge': discharge_10
        },
        'cycle100': {
            'voltage': voltage_100,
            'discharge': discharge_100
        }
    }

```

```python
# Function to compute Qdiffs for a given channel
def compute_Qdiffs_for_channel(channel_data):
    cycle10_discharge = channel_data['cycle10']['discharge']
    cycle100_discharge = channel_data['cycle100']['discharge']

    # Interpolating to compute Qdiffs
    f = interp1d(channel_data['cycle10']['voltage'], cycle10_discharge, bounds_error=False)
    Q_cycle10 = f(channel_data['cycle100']['voltage'])
    
    Qdiffs = cycle100_discharge - Q_cycle10
    return Qdiffs

# Computing Qdiffs for all channels
Qdiffs_all_channels = {}
for channel_num, channel_data in all_batch_adapted.items():
    Qdiffs_all_channels[channel_num] = compute_Qdiffs_for_channel(channel_data)

    
```

```python
# Function to overlay Qdiffs for all channels on a single plot
def overlay_Qdiffs_for_all_channels(Qdiffs_data, batch_adapted_data):
    

    # Plotting Qdiffs for all channels
    for channel_num, Qdiffs in Qdiffs_data.items():
        voltage_cycle100 = batch_adapted_data[channel_num]['cycle100']['voltage']
        plt.plot(Qdiffs, voltage_cycle100,  alpha=0.9)

    # Setting plot details
    plt.title('Qdiffs for All Channels',fontsize=20)
    plt.xlabel('Qdiffs (Ah)',fontsize=16)
    plt.ylabel('Voltage (V)',fontsize=16)
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Plotting the overlay of Qdiffs for all channels
overlay_Qdiffs_for_all_channels(Qdiffs_all_channels, all_batch_adapted)
```

```python

# Loading the provided cycle life data
cycle_life_data = pd.read_excel(r"C:\Users\GajeraRonakkumarIPCa\Desktop\Csv file Dataset\cycle life Csv data\cycle_life and channel_id.xlsx")

# Extracting channel IDs and their corresponding cycle lives
channel_ids = cycle_life_data['channel_id'].values
channel_cycle_lives = cycle_life_data['cycle_life'].values

```

```python
def get_Qs_adapted(data, cycle):
    vs, qds = data[cycle]['voltage'], data[cycle]['discharge']
    newF = interp1d(vs, qds)
    return newF, min(vs), max(vs)

def get_Qdiffs_adapted(data, nsamples=1000):
    # The curves for cycle 10 and 100
    Q10s = {key: get_Qs_adapted(value, 'cycle10') for key, value in data.items()}
    Q100s = {key: get_Qs_adapted(value, 'cycle100') for key, value in data.items()}
    
    Qdiffs_all = {}
    xvalues_all = {}
    
    for key in data.keys():
        q10 = Q10s[key]
        q100 = Q100s[key]
        
        # Find the overlapping range
        start = max(q10[1], q100[1])
        end   = min(q10[2], q100[2])
        x = np.linspace(start, end, nsamples)
        
        # Resample the curves and calculate Q100 - Q10
        curve10 = q10[0]
        curve100 = q100[0]
        
        Qdiffs_all[key] = curve100(x) - curve10(x)
        xvalues_all[key] = x

    return xvalues_all, Qdiffs_all

```

```python
# Proceeding with the visualization
import matplotlib.colors as mcolors


# Define your colormap
colors = ["darkblue", "green", "yellow", "darkred"]
cmap = mcolors.LinearSegmentedColormap.from_list("custom", colors, N=100)


# Extract xvalues and Qdiffs for all channels
xvalues_all, Qdiffs_all = get_Qdiffs_adapted(all_batch_adapted)

# Normalize cycle life for color mapping
norm = plt.Normalize(cycle_life_data['cycle_life'].min(), cycle_life_data['cycle_life'].max())

# Plot each curve of each channel with color-coding based on cycle life
for batch_key, xvalues in xvalues_all.items():
    cycle_life = cycle_life_data[cycle_life_data['channel_id'] == int(batch_key)]['cycle_life'].values[0]
    plt.plot(Qdiffs_all[batch_key], xvalues, color=cmap(norm(cycle_life)), label=f'Channel {batch_key} (Cycle Life: {cycle_life})')

# Add colorbar
sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])
cbar = plt.colorbar(sm)
cbar.set_label('Cycle Life', rotation=270, labelpad=15)

plt.grid(True)
plt.xlabel('Q100 - Q10',fontsize=16)
plt.ylabel('Voltage',fontsize=16)
#plt.legend(loc="upper left", bbox_to_anchor=(1,5))
plt.title('Color-coded  Discharge Capacity Differences vs. Voltage for All Channels',fontsize=20)
plt.tight_layout()
plt.show()
```

```python
def get_variances_adjusted(Qdiffs):
    return np.var(Qdiffs)


# Looping through each channel in all_batch_adapted to plot with the new color scheme
for channel_key in all_batch_adapted.keys():
    xvalues, Qdiffs = get_Qdiffs_adapted({channel_key: all_batch_adapted[channel_key]})
    fvalue = get_variances_adjusted(Qdiffs[channel_key])
    life = cycle_life_data[cycle_life_data['channel_id'] == int(channel_key)]['cycle_life'].values[0]
    
    # Plot the variance against cycle life for the channel with color coding
    plt.scatter(fvalue, life, color=cmap(norm(life)), label=f'Channel {channel_key} Data')

plt.xscale('log')
plt.yscale('log')
plt.grid(True)
plt.title('Variance of Discharge Capacity Differences vs. Cycle Life', fontsize=12)
plt.xlabel('Variance(Q100 - Q10)', fontsize=9)
plt.ylabel('Cycle Life', fontsize=9)

# Add colorbar
sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])
cbar = plt.colorbar(sm)
cbar.set_label('Cycle Life', rotation=270, labelpad=15)

plt.tight_layout()
plt.show()

```

```python
# Accumulating all variance and cycle life values across all channels
all_fvalues = []
all_lifes = []

for channel_key in all_batch_adapted.keys():
    xvalues, Qdiffs = get_Qdiffs_adapted({channel_key: all_batch_adapted[channel_key]})
    fvalues = get_variances_adjusted(Qdiffs[channel_key])
    lifes = cycle_life_data[cycle_life_data['channel_id'] == int(channel_key)]['cycle_life'].values[0]
    
    all_fvalues.append(fvalues)
    all_lifes.append(lifes)


log_all_fvalues = np.log(all_fvalues)
log_all_lifes = np.log(all_lifes)

overall_correlation_coefficient, _ = pearsonr(log_all_fvalues, log_all_lifes)

# Looping through each channel to add to the combined plot
# Looping through each channel to add to the combined plot with the new color scheme
for channel_key in all_batch_adapted.keys():
    xvalues, Qdiffs = get_Qdiffs_adapted({channel_key: all_batch_adapted[channel_key]})
    fvalues = get_variances_adjusted(Qdiffs[channel_key])
    lifes = cycle_life_data[cycle_life_data['channel_id'] == int(channel_key)]['cycle_life'].values[0]
    
    # Plot with color coding based on cycle life
    plt.scatter(np.log(fvalues), np.log(lifes), color=cmap(norm(lifes)), label=f'Channel {channel_key} Data')

plt.xlabel(f'Log of Variance(Q100 - Q10)')
plt.ylabel('Log of Cycle Life')
plt.title(f'Correlation coefficient: {overall_correlation_coefficient:.2f}')
plt.grid(True)

# Add colorbar
sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])
cbar = plt.colorbar(sm)
cbar.set_label('Cycle Life', rotation=270, labelpad=15)

plt.tight_layout()
plt.show()
```
